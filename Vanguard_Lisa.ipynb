## Testing Hypotheses 

```python
#H0: mu test = control
#H1: mu test != control
```

```python
#Convert "step_time" to hours, minutes, seconds

digital_footprints_kpi_time['hours'] = digital_footprints_kpi_time['step_time'].dt.components['hours']
digital_footprints_kpi_time['hours'].fillna(0, inplace = True)
digital_footprints_kpi_time['hours'] = digital_footprints_kpi_time['hours'].astype(int)

digital_footprints_kpi_time['minutes'] = digital_footprints_kpi_time['step_time'].dt.components['minutes']
digital_footprints_kpi_time['minutes'].fillna(0, inplace = True)
digital_footprints_kpi_time['minutes'] = digital_footprints_kpi_time['minutes'].astype(int)

digital_footprints_kpi_time['seconds'] = digital_footprints_kpi_time['step_time'].dt.components['seconds']
digital_footprints_kpi_time['seconds'].fillna(0, inplace = True)
digital_footprints_kpi_time['seconds'] = digital_footprints_kpi_time['seconds'].astype(int)
```

```python
digital_footprints_kpi_time['time_total_seconds'] = digital_footprints_kpi_time['hours']*3600 + \
                                                    digital_footprints_kpi_time['minutes'] *60 + \
                                                    digital_footprints_kpi_time['seconds']
```

```python
digital_footprints_kpi_time
```

```python
df_test = digital_footprints_kpi_time[digital_footprints_kpi_time["variation"]=="Test"][["process_step", 'time_total_seconds']]
test = df_test.pivot(columns='process_step', values='time_total_seconds')
test.fillna(0, inplace = True)
test.drop('confirm', axis=1, inplace = True)
test = test.rename(columns={'start': 'start_to_step_1', 'step_1': 'step_1_to_step_2', 'step_2': 'step_2_to_step_3', 'step_3': 'step_3_to_confirm'})
test
test
```

```python
digital_footprints_kpi_time.loc[digital_footprints_kpi_time['variation'] == 'Test']
```

```python
df_control = digital_footprints_kpi_time[digital_footprints_kpi_time["variation"]=="Control"][["process_step", 'time_total_seconds']]
control = df_control.pivot(columns='process_step', values='time_total_seconds')
control.fillna(0, inplace = True)
control.drop('confirm', axis=1, inplace = True)
control = control.rename(columns={'start': 'start_to_step_1', 'step_1': 'step_1_to_step_2', 'step_2': 'step_2_to_step_3', 'step_3': 'step_3_to_confirm'})
control
```

```python
result = st.ttest_ind(test, control, equal_var=False)
p_values = result[1]
```

```python
result[0]
```

```python
p_values
```

```python
alpha = 0.05
count = 0
for p_value in p_values:
    
    name_of_column = ['start_to_step_1', 'step_1_to_step_2', 'step_2_to_step_3', 'step_3_to_confirm']
   
    if p_value > alpha:
        print(f"{name_of_column[count]} time: We are not able to reject the null hypothesis")
    else:
        print(f"{name_of_column[count]} time: We reject the null hypothesis")
    count += 1
```

```python
#H0: mu test <= control
#H1: mu test > control
```

```python
result = st.ttest_ind(test, control, alternative = "greater")
p_values = result[1]
```

```python
p_values
```

```python
alpha = 0.05
count = 0
for p_value in p_values:
    
    name_of_column = ['start_to_step_1', 'step_1_to_step_2', 'step_2_to_step_3', 'step_3_to_confirm']
   
    if p_value > alpha:
        print(f"{name_of_column[count]} time: We are not able to reject the null hypothesis")
    else:
        print(f"{name_of_column[count]} time: We reject the null hypothesis")
    count += 1
```

# < Error Rates KPI

```python
#Copying df

digital_footprints_kpi_errors = digital_footprints_kpi.copy(deep=True)
```

```python
#Function for counting errors 

def check_errors(steps):
    steps_order = {
        "start": "step_1",
        "step_1": "step_2",
        "step_2": "step_3",
        "step_3": "confirm",
        "confirm": "",
    }
    
    # steps = digital_footprints_kpi_errors['process_steps']
    
    i_1 = 0
    i_2 = 1
    count_errors = 0
    for i in range(len(steps) - 1):
        if steps[i_2] != steps_order[steps[i_1]]:
            count_errors += 1
        i_1 += 1
        i_2 += 1
    
    if count_errors == 0 and len(steps) < 5:
        count_errors += 1

    return count_errors
```

```python
#Adding new column 'count_errors' using function 

digital_footprints_kpi_errors['count_errors'] = digital_footprints_kpi_errors.apply(lambda row: check_errors(row['process_steps']), axis = 1)
```

```python
digital_footprints_kpi_errors.head(5)
```

```python
#Checking data

digital_footprints_kpi_errors[['process_steps', 'count_errors']].loc[digital_footprints_kpi_errors['count_errors'] >= 4]
```

```python
#Adding new column 'count_steps' to count number of all steps

digital_footprints_kpi_errors['count_steps'] = digital_footprints_kpi_errors['process_steps'].str.len()
```

```python
#Defining new column 'error_rates_kpi'

digital_footprints_kpi_errors['error_rates_kpi'] = digital_footprints_kpi_errors['count_errors']/digital_footprints_kpi_errors['count_steps']
digital_footprints_kpi_errors['error_rates_kpi'] = digital_footprints_kpi_errors['error_rates_kpi'].round(2)
```

# < Test other Hypothesis 

### You might want to test whether the average age of clients engaging with the new process is the same as those engaging with the old process.

```python
#H0: mu test = control
#H1: mu test != control
```

```python
df_test_age = client_profiles[client_profiles["variation"]=="Test"]["clnt_age"]
df_control_age = client_profiles[client_profiles["variation"]=="Control"]["clnt_age"]
```

```python
print(f'New version: {df_test_age.mean()}')
print(f'Old version: {df_control_age.mean()}')
```

```python
result = st.ttest_ind(df_test_age, df_control_age , equal_var=True)
p_value = result[1]
p_value
```

```python
alpha = 0.05

if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypothesis")
```

```python
#Used for checking

#In order to calculate test statistic we need

#sample mean
mean = df_test_age.mean()

#standard deviation of sample
s = df_test_age.std(ddof=1)

#sample size
n = len(df_test_age)

#hypothesized population mean
mu = df_control_age.mean()

stat = (mean - mu)/(s/np.sqrt(n))
print(f'stat: {stat}')

p_value = st.t.sf(abs(stat), n-1)*2
print(f'p_value: {p_value}')
```

```python
if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```
